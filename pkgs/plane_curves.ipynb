```python
from sympy import plot_parametric, symbols
```

```python
t = symbols("t")
```

```python
def x(t):
    return t


def y(t):
    return t ** 2
```

```python
plot_parametric((x(t), y(t), (t, 0, 5)))
```

```python
def k(t):
    d = x(t).diff(t, 1) * y(t).diff(t, 2) - y(t).diff(t, 1) * x(t).diff(t, 2)
    v = (x(t).diff(t, 1) ** 2 + y(t).diff(t, 1) ** 2) ** 0.5
    return (d / v).simplify()
```

```python
k(t)
```
