```python
from sympy import plot_parametric, symbols, cos, sin, exp
```

```python
t = symbols('t')
```

```python
def x(t):
    return cos(t)

def y(t):
    return sin(t)
```

```python
plot_parametric((x(t), y(t), (t, 0, 6.28)))
```

```python
def v(t):
    return ((x(t).diff(t, 1)**2 + y(t).diff(t, 1)**2)**0.5).simplify()
```
